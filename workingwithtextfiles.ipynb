```python
name = 'Fred'

print('his name is {var}.'.format(var=name))

print(f'his name is {name}.')
```

```python
print(f'his name is {name!r}')
```

```python
library = [('author', 'topic', 'pages'), ('feyman', 'physics', 95), ('hamilton', 'mythology', 144)]

for book in library:
    print(f'{book[0]:{10}} {book[1]:{8}} {book[2]:{7}}')
```

```python
for book in library:
    print(f'{book[0]:{10}} {book[1]:{8}} {book[2]:.>{7}}')
```

```python
with open('test.txt', 'r') as txt:
    first_line = txt.readlines()[0]

print(first_line)
```

```python
with open('test.txt', 'r') as txt:
    for line in txt:
        print(line)
```

```python
with open('test.txt', 'r') as txt:
    for line in txt:
        print(line, end='')
```
